```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/df_numeric (3).csv")
X=df
y=df['ytest']
df.shape
```

```python
############################## Scaling of Values ##########################
scaler=StandardScaler()
scaler.fit(X)

scaled_data=scaler.transform(X)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=96)
pca.fit(scaled_data)

x_pca=pca.transform(scaled_data)
x_pca
```

```python
train_df= pd.DataFrame(x_pca)
train_df
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
import xgboost as xgb
```

```python
######################################### SVR #################################################

X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=0.2, random_state=42)

# Scale the features using standard scaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

regressor = SVR(kernel='linear')
regressor.fit(X_train, y_train)

# Make predictions on the testing dataset
y_pred1 = regressor.predict(X_test)
y_pred2 = regressor.predict(X_train)

# Evaluate the model using R2 score and mean squared error
r21 = r2_score(y_train,y_pred2)
r2 = r2_score(y_test, y_pred1)
mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
print("R square: ",r21)
print("R2 square test:", r2)
print("Mean Squared Error:", mse)
print("RMSE: ",rmse)

plt.scatter(y_train, y_pred2,  facecolors='none', edgecolors='blue')
plt.scatter(y_test, y_pred1, color='red', marker='+')
plt.legend(["Training Set", "Testing Set"])
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
####################################### Linear Regression #######################################

X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=0.2, random_state=42)

# Scale the features using standard scaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

model = LinearRegression()
model.fit(X_train, y_train)

# Test the model on the testing dataset and calculate the mean squared error
y_pred1 = model.predict(X_test)
y_pred2 = model.predict(X_train)


r21 = r2_score(y_train,y_pred2)
r2 = r2_score(y_test, y_pred1)
mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
print("R square: ",r21)
print("R2 square test:", r2)
print("Mean Squared Error:", mse)
print("RMSE: ",rmse)

plt.scatter(y_train, y_pred2,  facecolors='none', edgecolors='blue')
plt.scatter(y_test, y_pred1, color='red', marker='+')
plt.legend(["Training Set", "Testing Set"])
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
########################################## XGBoost ##########################################

X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=0.2, random_state=42)

# Define the XGBoost model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=3)

# Train the XGBoost model
xgb_model.fit(X_train, y_train)

# Make predictions on the testing dataset
y_pred1 = xgb_model.predict(X_test)
y_pred2 = xgb_model.predict(X_train)

# Evaluate the model using R2 score and mean squared error
r21 = r2_score(y_train,y_pred2)
r2 = r2_score(y_test, y_pred1)
mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
print("R square: ",r21)
print("R2 square test:", r2)
print("Mean Squared Error:", mse)
print("RMSE: ",rmse)

plt.scatter(y_train, y_pred2,  facecolors='none', edgecolors='blue')
plt.scatter(y_test, y_pred1, color='red', marker='+')
plt.legend(["Training Set", "Testing Set"])
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
###################################### LARS ################################

from sklearn.linear_model import Lars

X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

lars = Lars()
lars.fit(X_train, y_train)

y_pred1 = lars.predict(X_test)
y_pred2 = lars.predict(X_train)

mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred1)
r2tr = r2_score(y_train,y_pred2)
print("R square score: ",r2tr)
print("R square test score: ",r2)
print('Mean squared error:', mse)
print("RMSE: ", rmse)

plt.scatter(y_train, y_pred2,  facecolors='none', edgecolors='blue')
plt.scatter(y_test, y_pred1, color='red', marker='+')
plt.legend(["Training Set", "Testing Set"])
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
################################ LGBM ##########################################D
from lightgbm import LGBMRegressor

X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

lgbm = LGBMRegressor()
lgbm.fit(X_train, y_train)

y_pred1 = lgbm.predict(X_test)
y_pred2 = lgbm.predict(X_train)

mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred1)
r2tr = r2_score(y_train,y_pred2)
print("R square score: ",r2tr)
print("R square test score: ",r2)
print('Mean squared error:', mse)
print("RMSE: ", rmse)

plt.scatter(y_train, y_pred2,  facecolors='none', edgecolors='blue')
plt.scatter(y_test, y_pred1, color='red', marker='+')
plt.legend(["Training Set", "Testing Set"])
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
